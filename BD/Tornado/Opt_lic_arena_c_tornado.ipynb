```python
%%time

w=0
df_dict = {}
    
wb = xw.Book('./FCL_Arena_C.xlsx')
sht1 = wb.sheets['Perfiles']
sht2= wb.sheets['FCL Total']
sht3=wb.sheets['FCL Basica']
sht4=wb.sheets['FCL Resumen']


for a in range(7):
    
    capex_ph=capex_p[a]
    opex_ph=opex_p[a]
    produc_ph=produc_p[a]

    
    ### riesgo (2)
    for b in range(2):
        AR_b=AR[b]
        BR_b=BR[b]

        ### Incremental WOs (3)

        for c in range(2):

            WO_c=WO[c]
            NW_d=NW[c]

        ## CIRA  (2)

            for e in range(2):

                # WOs perfil

                oil_C_WO_AR=WO_C_AR_y.iloc[:,1]*opt_C[e]*AR_b*WO_c
                fluido_C_WO_AR=WO_C_AR_y.iloc[:,2]*opt_C[e]*AR_b*WO_c
                gas_C_WO_AR=WO_C_AR_y.iloc[:,3]*opt_C[e]*AR_b*WO_c
                inyeccion_C_WO_AR=WO_C_AR_y.iloc[:,4]*opt_C[e]*AR_b*WO_c


                oil_C_WO_BR=WO_C_BR_y.iloc[:,1]*opt_C[e]*BR_b*WO_c
                fluido_C_WO_BR=WO_C_BR_y.iloc[:,2]*opt_C[e]*BR_b*WO_c
                gas_C_WO_BR=WO_C_BR_y.iloc[:,3]*opt_C[e]*BR_b*WO_c
                inyeccion_C_WO_BR=WO_C_BR_y.iloc[:,4]*opt_C[e]*BR_b*WO_c                   


                 # WOs capex (perforacion)

                WOs_p_C=(P_WO_C_AR_v.iloc[:,1]*costo_pozo_WO*opt_C[e]*AR_b*WO_c +
                          P_WO_C_BR_v.iloc[:,1]*costo_pozo_WO*opt_C[e]*BR_b*WO_c)


                 # WOs Facilidades

                WOs_f_C=(bd_capex_WO_C_AR_y.iloc[:,1]*opt_C[e]*AR_b*WO_c + 
                          bd_capex_WO_C_BR_y.iloc[:,1]*opt_C[e]*BR_b*WO_c)


                # Nw perfil
                oil_C_NW_BR=NW_C_BR_y.iloc[:,1]*opt_C[e]*BR_b*NW_d
                fluido_C_NW_BR=NW_C_BR_y.iloc[:,2]*opt_C[e]*BR_b*NW_d
                gas_C_NW_BR=NW_C_BR_y.iloc[:,3]*opt_C[e]*BR_b*NW_d
                inyeccion_C_NW_BR=NW_C_BR_y.iloc[:,4]*opt_C[e]*BR_b*NW_d


                p_prod_C_BR=P_NW_C_BR_y.iloc[:,1]*opt_C[e]*BR_b*NW_d
                p_iny_C_BR=P_NW_C_BR_y.iloc[:,2]*opt_C[e]*BR_b*NW_d



                oil_C_NW_AR=NW_C_AR_y.iloc[:,1]*opt_C[e]*AR_b*NW_d
                fluido_C_NW_AR=NW_C_AR_y.iloc[:,2]*opt_C[e]*AR_b*NW_d
                gas_C_NW_AR=NW_C_AR_y.iloc[:,3]*opt_C[e]*AR_b*NW_d
                inyeccion_C_NW_AR=NW_C_AR_y.iloc[:,4]*opt_C[e]*AR_b*NW_d


                p_prod_C_AR=P_NW_C_AR_y.iloc[:,1]*opt_C[e]*AR_b*NW_d
                p_iny_C_AR=P_NW_C_AR_y.iloc[:,2]*opt_C[e]*AR_b*NW_d


                # Nw capex (perforacion) - NUMERO DE POZOS

                NW_p_C=(C_NW_C_BR_y.iloc[:,1]*costo_pozo_arena_c_p*opt_C[e]*BR_b*NW_d+
                         C_NW_C_BR_y.iloc[:,2]*costo_pozo_arena_c_i*opt_C[e]*BR_b*NW_d+
                         C_NW_C_AR_y.iloc[:,1]*costo_pozo_arena_c_p*opt_C[e]*AR_b*NW_d+
                         C_NW_C_AR_y.iloc[:,2]*costo_pozo_arena_c_i*opt_C[e]*AR_b*NW_d)


                # Nw Facilidades

                NW_f_C=(bd_capex_NW_C_BR_y.iloc[:,1]*opt_C[e]*BR_b*NW_d+
                         bd_capex_NW_C_AR_y.iloc[:,1]*opt_C[e]*AR_b*NW_d)

                # Nw Abandono

                NW_A_C=(P_NW_C_BR_y.iloc[:,3]*costo_abandono_c*opt_C[e]*BR_b*NW_d+
                         P_NW_C_AR_y.iloc[:,3]*costo_abandono_c*opt_C[e]*AR_b*NW_d)


                ## infanta  (2)

                for h in range(2):

                    oil_I_WO_AR=WO_I_AR_y.iloc[:,1]*opt_I[h]*AR_b*WO_c
                    fluido_I_WO_AR=WO_I_AR_y.iloc[:,2]*opt_I[h]*AR_b*WO_c
                    gas_I_WO_AR=WO_I_AR_y.iloc[:,3]*opt_I[h]*AR_b*WO_c
                    inyeccion_I_WO_AR=WO_I_AR_y.iloc[:,4]*opt_I[h]*AR_b*WO_c


                    oil_I_WO_BR=WO_I_BR_y.iloc[:,1]*opt_I[h]*BR_b*WO_c
                    fluido_I_WO_BR=WO_I_BR_y.iloc[:,2]*opt_I[h]*BR_b*WO_c
                    gas_I_WO_BR=WO_I_BR_y.iloc[:,3]*opt_I[h]*BR_b*WO_c
                    inyeccion_I_WO_BR=WO_I_BR_y.iloc[:,4]*opt_I[h]*BR_b*WO_c                   


                     # WOs capex (perforacion)

                    WOs_p_I=(P_WO_I_AR_v.iloc[:,1]*costo_pozo_WO*opt_I[h]*AR_b*WO_c +
                              P_WO_I_BR_v.iloc[:,1]*costo_pozo_WO*opt_I[h]*BR_b*WO_c)


                     # WOs Facilidades

                    WOs_f_I=(bd_capex_WO_I_AR_y.iloc[:,1]*opt_I[h]*AR_b*WO_c + 
                              bd_capex_WO_I_BR_y.iloc[:,1]*opt_I[h]*BR_b*WO_c)


                    # Nw perfil
                    oil_I_NW_BR=NW_I_BR_y.iloc[:,1]*opt_I[h]*BR_b*NW_d
                    fluido_I_NW_BR=NW_I_BR_y.iloc[:,2]*opt_I[h]*BR_b*NW_d
                    gas_I_NW_BR=NW_I_BR_y.iloc[:,3]*opt_I[h]*BR_b*NW_d
                    inyeccion_I_NW_BR=NW_I_BR_y.iloc[:,4]*opt_I[h]*BR_b*NW_d


                    p_prod_I_BR=P_NW_I_BR_y.iloc[:,1]*opt_I[h]*BR_b*NW_d
                    p_iny_I_BR=P_NW_I_BR_y.iloc[:,2]*opt_I[h]*BR_b*NW_d



                    oil_I_NW_AR=NW_I_AR_y.iloc[:,1]*opt_I[h]*AR_b*NW_d
                    fluido_I_NW_AR=NW_I_AR_y.iloc[:,2]*opt_I[h]*AR_b*NW_d
                    gas_I_NW_AR=NW_I_AR_y.iloc[:,3]*opt_I[h]*AR_b*NW_d
                    inyeccion_I_NW_AR=NW_I_AR_y.iloc[:,4]*opt_I[h]*AR_b*NW_d


                    p_prod_I_AR=P_NW_I_AR_y.iloc[:,1]*opt_I[h]*AR_b*NW_d
                    p_iny_I_AR=P_NW_I_AR_y.iloc[:,2]*opt_I[h]*AR_b*NW_d


                    # Nw capex (perforacion) - NUMERO DE POZOS

                    NW_p_I=(C_NW_I_BR_y.iloc[:,1]*costo_pozo_arena_c_p*opt_I[h]*BR_b*NW_d+
                             C_NW_I_BR_y.iloc[:,2]*costo_pozo_arena_c_i*opt_I[h]*BR_b*NW_d+
                             C_NW_I_AR_y.iloc[:,1]*costo_pozo_arena_c_p*opt_I[h]*AR_b*NW_d+
                             C_NW_I_AR_y.iloc[:,2]*costo_pozo_arena_c_i*opt_I[h]*AR_b*NW_d)


                    # Nw Facilidades

                    NW_f_I=(bd_capex_NW_I_BR_y.iloc[:,1]*opt_I[h]*BR_b*NW_d+
                             bd_capex_NW_I_AR_y.iloc[:,1]*opt_I[h]*AR_b*NW_d)

                    # Nw Abandono

                    NW_A_I=(P_NW_I_BR_y.iloc[:,3]*costo_abandono_i*opt_I[h]*BR_b*NW_d+
                             P_NW_I_AR_y.iloc[:,3]*costo_abandono_i*opt_I[h]*AR_b*NW_d)                                    


                    #### (n)

                    for i in range(n):

                        ## adecuacion de las base de datos 


                        ### perfiles

                        ## WOs
                        oil_WOs=(oil_C_WO_AR + oil_C_WO_BR +oil_I_WO_AR + oil_I_WO_BR)*produc_ph

                        fluido_WOs=(fluido_C_WO_AR + fluido_C_WO_BR +fluido_I_WO_AR + fluido_I_WO_BR)*produc_ph

                        gas_WOs=(gas_C_WO_AR + gas_C_WO_BR +gas_I_WO_AR + gas_I_WO_BR)*produc_ph

                        inyeccion_WOs=(inyeccion_C_WO_AR +inyeccion_C_WO_BR +
                                       inyeccion_I_WO_AR +inyeccion_I_WO_BR)*produc_ph


                        ## NW

                        oil_NW=(oil_C_NW_BR+oil_C_NW_AR+oil_I_NW_BR+oil_I_NW_AR)*produc_ph

                        fluido_NW=(fluido_C_NW_BR+fluido_C_NW_AR+fluido_I_NW_BR+fluido_I_NW_AR)*produc_ph

                        gas_NW=(gas_C_NW_BR+gas_C_NW_AR+gas_I_NW_BR+gas_I_NW_AR)*produc_ph

                        inyeccion_NW=(inyeccion_C_NW_BR+inyeccion_C_NW_AR+
                                      inyeccion_I_NW_BR+inyeccion_I_NW_AR)*produc_ph

                        # pozos perfiles

                        productor=(p_prod_C_BR+p_prod_C_AR+p_prod_I_BR+p_prod_I_AR)

                        inyector=(p_iny_C_BR+p_iny_C_AR++p_iny_I_BR+p_iny_I_AR)

                        # capex (perforacion)


                        perforacion=(WOs_p_C+NW_p_C+WOs_p_I+NW_p_I)*capex_ph


                        # facilidades

                        facilidades=(WOs_f_C+NW_f_C+WOs_f_I+NW_f_I)*capex_ph


                        # abandono
                        abandono=(NW_A_C+NW_A_I)*capex_ph

                        ### incremental

                        # Perfil
                           # WOs
                        sht1.range('K69').options(transpose=False).value=oil_WOs.values
                        sht1.range('K70').options(transpose=False).value=gas_WOs.values
                        sht1.range('K71').options(transpose=False).value=(fluido_WOs-oil_WOs).values
                        sht1.range('K72').options(transpose=False).value=inyeccion_WOs.values

                           #NW

                        sht1.range('K85').options(transpose=False).value=oil_NW.values
                        sht1.range('K86').options(transpose=False).value=gas_NW.values
                        sht1.range('K87').options(transpose=False).value=(fluido_NW-oil_NW).values
                        sht1.range('K88').options(transpose=False).value=inyeccion_NW.values


                        # WO pozo perfil
                        sht1.range('K90').options(transpose=False).value=productor.values
                        sht1.range('K91').options(transpose=False).value=inyector.values


                        # capex (perforacion)
                        sht2.range('F15').options(transpose=False).value=(perforacion/1000).values



                        # facilidades
                        sht2.range('F16').options(transpose=False).value=(facilidades/1000).values


                        # abandono pozos
                        sht2.range('F22').options(transpose=False).value=(abandono/1000).values
                        
                        
                        
                        oil_opex=decision_opex[decision_opex['decision']== D[0]].iloc[:,1]
                        Gas_Prod_opex=decision_opex[decision_opex['decision']== D[0]].iloc[:,2]
                        Gross_Fluid_opex=decision_opex[decision_opex['decision']== D[0]].iloc[:,3]
                        Water_Injected_opex_i=decision_opex[decision_opex['decision']== D[0]].iloc[:,4]
                        Producer_Well_opex=decision_opex[decision_opex['decision']== D[0]].iloc[:,5]
                        Water_Injection_Well_opex=decision_opex[decision_opex['decision']== D[0]].iloc[:,6]

                        ## opex

                        sht1.range('N43').options(transpose=True).value=(oil_opex.values)*opex_ph
                        sht1.range('P43').options(transpose=True).value=(Gas_Prod_opex.values)*opex_ph
                        sht1.range('Q43').options(transpose=True).value=(Gross_Fluid_opex.values)*opex_ph
                        sht1.range('R43').options(transpose=True).value=(Water_Injected_opex_i.values)*opex_ph
                        sht1.range('U43').options(transpose=True).value=(Producer_Well_opex.values)*opex_ph
                        sht1.range('V43').options(transpose=True).value=(Water_Injection_Well_opex.values)*opex_ph
                            
                            


                        ## cambio de brent 
                        sht4.range('C10').value=E_Brent[i]

                        if E_Brent[i]<=30:
                            PP=0.52+0

                        elif E_Brent[i]==35:

                            PP=0.52+0.015

                        elif E_Brent[i]==40:

                            PP=0.52+0.035

                        elif E_Brent[i]==45:

                            PP=0.52+0.05

                        elif E_Brent[i]==50:

                            PP=0.52+0.05

                        else:
                            PP=0.52+0.07



                        # abandono pozos
                        sht3.range('I6').options(transpose=False).value=PP


                        wb.app.calculate()

                        VPN_Basica= sht4.range('E10').value
                        VPN_Total = sht4.range('E18').value
                        VPN_incremental= sht4.range('E26').value
                        VPN_Basica_porta=sht4.range('E8').value
                        VPN_incremental_porta=sht4.range('E24').value

                        EFI_INCREMENTAL=sht4.range('F26').value
                        capex=sht2.range('E21').value
                        opex_basica=sht3.range('E26').value
                        opex_incremental=sht2.range('E26').value
                        volumen_B_LE=sht4.range('i62').value
                        volumen_I_LE=sht4.range('i74').value
                        volumen_T_LE=sht4.range('i86').value


                        volumen_B_LE_p=sht4.range('i60').value
                        volumen_I_LE_p=sht4.range('i72').value
                        volumen_T_LE_p=sht4.range('i84').value

                        ## limite economico

                        LE_year_p= sht4.range('K16').value
                        LE_year_b= sht4.range('K18').value

                        ### 

                        lc_p=sht4.range('E42').value
                        lc_b=sht4.range('E44').value
                        dc_p=sht4.range('F42').value
                        dc_b=sht4.range('F44').value

                        ######

                        opex_ing_p=sht4.range('M16').value
                        capex_ing_p=sht4.range('N16').value
                        impuesto_ing_p=sht4.range('O16').value
                        ingreso_ing_p=sht4.range('P16').value
                        abandono_ing_p=sht4.range('Q16').value

                        opex_ing_b=sht4.range('M18').value
                        capex_ing_b=sht4.range('N18').value
                        impuesto_ing_b=sht4.range('O18').value
                        ingreso_ing_b=sht4.range('P18').value
                        abandono_ing_b=sht4.range('Q18').value



                        row_ = {'ID':w,
                            'E_Brent':E_Brent[i],
                            'PP':PP,
                            'BR':BR[b],
                            'AR':AR[b],
                            'WO':WO[c],
                            'NW':NW[c],   
                            'opt_C':opt_C[e],
                            'opt_I':opt_I[h], 
                            'capex_h':capex_p[x],
                            'opex_h':opex_p[x],
                            'produc_h':produc_p[x],
                            'VPN_Basica_porta':VPN_Basica_porta,
                            'VPN_incremental_porta':VPN_incremental_porta,
                            'VPN_Basica':VPN_Basica,
                            'VPN_incremental':VPN_incremental,
                            'VPN_Total':VPN_Total,
                            'EFI_INCREMENTAL':EFI_INCREMENTAL,
                            'capex':capex,
                            'opex_basica':opex_basica,
                            'opex_incremental':opex_incremental,
                            'volumen_B_LE':volumen_B_LE/1000,
                            'volumen_I_LE':volumen_I_LE/1000,
                            'volumen_T_LE':volumen_T_LE/1000,
                            'volumen_B_LE_p':volumen_B_LE_p/1000,
                            'volumen_I_LE_p':volumen_I_LE_p/1000,
                            'volumen_T_LE_p':volumen_T_LE_p/1000,
                            'LE_year_p':LE_year_p,
                            'LE_year_b':LE_year_b,
                            'lc_p':lc_p,
                            'lc_b':lc_b,
                            'dc_p':dc_p,
                            'dc_b':dc_b,
                            'opex_ing_p':opex_ing_p,
                            'capex_ing_p':capex_ing_p,
                            'impuesto_ing_p':impuesto_ing_p,
                            'ingreso_ing_p':ingreso_ing_p,
                            'abandono_ing_p':abandono_ing_p,
                            'opex_ing_b':opex_ing_b,
                            'capex_ing_b':capex_ing_b,
                            'impuesto_ing_b':impuesto_ing_b,
                            'ingreso_ing_b':ingreso_ing_b,
                            'abandono_ing_b':abandono_ing_b,
                        }


                        df_dict[w]=row_

                        w+=1

```
